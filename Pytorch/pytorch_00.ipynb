## Creating Tensors

```python
scalar = torch.tensor(7)
scalar
```

To get the numbe inside the object


```python
scalar.item()
```

```python
vector = torch.tensor([2,4])
vector
```

```python
print('vector.ndim : ',vector.ndim, 'it gives me the dimension of the tensor: 0 scalar, 1 vector, 2 matrix, ...')
print('vector.shape: ', vector.shape, 'it tells me about how the elements of the tensor are organized')
```

```python
matrix = torch.tensor([ [1,2],[3,4]])
matrix
```

**You can continue with this idea and create a tensor of any dimension**

Random tensors

```python
rand_tensor = torch.rand((3,4))
rand_tensor
```

Tensor of zeros and ones

```python
zeros_tensor = torch.zeros((4,4))
zeros_tensor
```

```python
ones_tensor = torch.ones((2,3))
ones_tensor
```

Creating a range tensor

```python
range_tensor = torch.arange(0,10,2)
range_tensor
```

You can use the shape of another tensor to create a tensor of zeros or ones with the same shape

```python
tensor_alike = torch.zeros_like(range_tensor)
tensor_alike
```

```python
tensor_alike_ones = torch.ones_like(range_tensor)
tensor_alike_ones
```

**Creating tensors with defined datatype**

```python
float_32_tensor = torch.tensor([1,2,3],
                               dtype=None, # defaults to None, which is torch.float32 or whatever datatype is passed
                               device=None, # defaults to None, which uses the default tensor type,  what device is the tensor stored on? (usually GPU or CPU)
                               requires_grad=False) # if True, operations  performed on the tensor are recorded
```

#### PyTorch tensors & NumPy

The two main methods you'll want to use for NumPy to PyTorch (and back again) are:




*   torch.from_numpy(ndarray) - NumPy array -> PyTorch tensor.
*   torch.Tensor.numpy() - PyTorch tensor -> NumPy array.




```python
import numpy as np
```

```python
array = np.arange(0,20,4)
tensor = torch.from_numpy(array)

array, tensor
```
